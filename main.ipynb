## Removing the Contradictions

There were some contradictions in the dataset where the answers given to some questions and their assigned classes were not making sense, so we removed those rows.

```python
def handle_contradictions(df, ct_mask, contradictions_accumulated=None):
    
    contradictions_accumulated = pd.concat([contradictions_accumulated, df[ct_mask]])
    
    df = df.drop(df[ct_mask].index)

    return df, contradictions_accumulated
```

```python
#An empty df to store all the contradictions
all_contradictions = pd.DataFrame()
```

Q12_14 states "I would never buy an EV". If the person answered "Yes (1)" to this question, the assigned class being "A (If I had an EV, it would be my only car)" seems like a logic error.

```python
contradictions_mask = ((df['Q16'] == 0) & (df['Q12_14'] == 1) | (df['Q16'] == 2) & (df['Q12_14'] == 0))
df, all_contradictions = handle_contradictions(df, contradictions_mask, all_contradictions)
```

## Calculating Metrics

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, matthews_corrcoef

import pandas as pd

def evaluate_classifier(classifier, X_train, y_train, X_test, y_test):
    
    classifier.fit(X_train, y_train)
    
    y_pred = classifier.predict(X_test)
    y_probs = classifier.predict_proba(X_test)
    
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')
    roc_auc = roc_auc_score(y_test, y_probs, multi_class='ovr')
    mcc = matthews_corrcoef(y_test, y_pred)
    
    metrics_df = pd.DataFrame({
        'Metric': ['Accuracy', 'Precision', 'Recall', 'F1 Score', 'ROC-AUC', 'MCC'],
        'Score': [accuracy, precision, recall, f1, roc_auc, mcc]
    })
    
    print("Metrics DataFrame:")
    display(metrics_df)
```

## ROC for All Algorithms

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

classifiers_list = [rf_classifier, stacking_clf, gb_classifier, tree_classifier, bagging_classifier, knn_classifier, naive_bayes_classifier, xgb_classifier, svm_classifier, extra_trees_classifier]
classifier_names = ["Random Forest" , "Stacking" , "Gradient Boosting" , "Decision Tree", "Bagging", "kNN", "Naive Bayes", "XGBoost", "SVM", "Extra Trees"]

fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(16, 6))

# Flattening the axes to iterate over them
axes = axes.flatten()

for classifier, name, ax in zip(classifiers_list, classifier_names, axes):
    classifier.fit(X_train, y_train)
    y_probs = classifier.predict_proba(X_test)
    n_classes = len(classifier.classes_)

    for i in range(n_classes):
        fpr, tpr, _ = roc_curve((y_test == classifier.classes_[i]).astype(int), y_probs[:, i])
        roc_auc = auc(fpr, tpr)
        ax.plot(fpr, tpr, lw=2, label=f'Class {classifier.classes_[i]} (AUC = {roc_auc:.2f})')

    ax.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_title(name)
    ax.legend(loc='lower right')

plt.tight_layout()
plt.show()
```

## Feature Selection (RFE)

```python
from sklearn.feature_selection import RFE

def feature_sel_RFE(classifier, importance_classifier, X_train, y_train, X_test, y_test):
        
    rfe_selected_features = []    
    rfe_accuracy_list = []
    rfe_k = []

    for k in range(1, 75):
        RFE_selector = RFE(estimator=importance_classifier, n_features_to_select=k, step=10)
        RFE_selector.fit(X_train, y_train)
        
        selected_mask = RFE_selector.support_

        selected_indices = [i for i, selected in enumerate(selected_mask) if selected]
        selected_features = X_train.columns[selected_indices]

        rfe_selected_features.append(selected_features.tolist())

        sel_X_train = RFE_selector.transform(X_train)
        sel_X_test = RFE_selector.transform(X_test)

        classifier.fit(sel_X_train, y_train)
        RFE_preds = classifier.predict(sel_X_test)

        accuracy = accuracy_score(y_test, RFE_preds)
        rfe_accuracy_list.append(accuracy)
        rfe_k.append(k)

    best_accuracy_index = np.argmax(rfe_accuracy_list)

    best_k = rfe_k[best_accuracy_index]
    best_accuracy = rfe_accuracy_list[best_accuracy_index]

    pd.set_option('display.max_rows', None)
    pd.set_option('display.max_colwidth', None) 
 
    results_df = pd.DataFrame({
    'k': rfe_k,
    'accuracy': rfe_accuracy_list,
    'selected_features': rfe_selected_features
    })

    display(results_df.iloc[[best_k-1]])
```

## Cross Validation

```python
from sklearn.ensemble import RandomForestClassifier

selected_classifier = RandomForestClassifier(n_estimators=250,random_state = 42)
```

```python
RFE_selector = RFE(estimator=selected_classifier, n_features_to_select=66, step=10)
RFE_selector.fit(X_train, y_train)

sel_X_train = RFE_selector.transform(X_train)
sel_X_test = RFE_selector.transform(X_test)

selected_classifier.fit(sel_X_train, y_train)

evaluate_classifier(selected_classifier, sel_X_train, y_train, sel_X_test, y_test)
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)
 
accuracy_scores = []

fold_number = 1
for train_index, test_index in kf.split(X): 
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    RFE_selector = RFE(estimator=selected_classifier, n_features_to_select=66, step=10)
    RFE_selector.fit(X_train, y_train)

    sel_X_train = RFE_selector.transform(X_train)
    sel_X_test = RFE_selector.transform(X_test)

    selected_classifier.fit(sel_X_train, y_train)

    RFE_preds = selected_classifier.predict(sel_X_test)

    accuracy = accuracy_score(y_test, RFE_preds)
    print(f"Accuracy for fold {fold_number}: {accuracy:.3f}")
    accuracy_scores.append(accuracy)
    fold_number += 1


print(f'Cross Validation Mean Accuracy: {np.mean(accuracy_scores):.3f}')
```
